### Handle the missing values

```python
from sklearn.impute import SimpleImputer
```

```python
df.describe()
```

```python
df["bmi"]=df["bmi"].replace(np.NaN,df["bmi"].mean())
print(df["bmi"])
```

### Handling the categorical values

```python
cat_cols =['gender','ever_married','work_type','Residence_type','smoking_status']
from sklearn.preprocessing import LabelEncoder
for each_col in cat_cols:
    Le=LabelEncoder()
    df[each_col]=Le.fit_transform (df[each_col])
```

```python
df
```

### Split the dataset into training and testing 

```python
x=df[['gender','age','hypertension','heart_disease','ever_married','work_type','Residence_type','avg_glucose_level','bmi',
      'smoking_status']].values
y=df[['stroke']].values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size =0.8)
```

```python
y_test.shape
```

### Import and perform linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr =LinearRegression()
lr.fit(x,y)
print("Our model has been Trained!!")
```

### Predict the values

```python
y_pred =lr.predict(x_test)
y_pred.round(2)
```
